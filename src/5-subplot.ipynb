### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Create a figure with a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Data for line chart
x = np.linspace(0, 10, 100)
y = np.sin(x)
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X Values')
axs[0, 0].set_ylabel('Y Values')
axs[0, 0].grid(True)

# Data for bar chart
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Data for scatter plot
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='blue')
for i, j in zip(x_scatter, y_scatter):
    axs[1, 0].annotate(f'({i}, {j})', (i, j), textcoords="offset points", xytext=(5,5), ha='right')
    axs[1, 0].set_title('Scatter Plot')
    axs[1, 0].set_xlabel('X Values')
    axs[1, 0].set_ylabel('Y Values')
    axs[1, 0].grid(True)

    # Data for pie chart
    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'purple'])
    axs[1, 1].set_title('Pie Chart')

    # Adjust layout
    plt.tight_layout()

# Display the plots
    plt.show()
```
